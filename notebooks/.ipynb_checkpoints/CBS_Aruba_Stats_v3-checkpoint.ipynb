# Population Structure and Vulnerability: A Demographic Analysis of Aruba (2016–2023)

This notebook explores CBS demographic data for Aruba, revealing critical trends in population aging and migration dependency that shape the island's future. By examining age cohorts, natural increase, and net migration patterns, we identify structural vulnerabilities that have implications for healthcare, the workforce, and economic sustainability.

This analysis is part of a broader investigation into Aruba's dependencies and vulnerabilities, examining health data, economic indicators, and social structures alongside demographic trends.

## Objective
This notebook visualizes the total population of males and females in Aruba
between 2015 and 2023 using data from the Central Bureau of Statistics.
The goal is to understand long-term demographic trends before analyzing
year-to-year population changes.

## Data Source
- **Source: Central Bureau of Statistics Aruba**
- **Dataset:** Key demographic aspects.xlsx
- **Period:** 2015–2023(Note: 2015 data is incomplete and excluded from analysis)
- **Variables:** Absolute head counts of males vs females

## Analytical purpose
Establish baseline population trends before examining population change
and migration drivers in subsequent notebooks.

## Overview

This notebook is structured as follows:

1. **Setup & Configuration** — Import libraries and establish file paths
2. **Data Loading & Exploration** — Load CBS datasets and examine their structure
3. **Analysis & Visualization** — Transform data, calculate trends, and visualize findings

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Import Class
from pathlib import Path

# Set Path to current working directory
cwd = Path.cwd()

# Set filesystem paths
ROOT = cwd.parent if cwd.name == "notebooks" else cwd

# Establish paths
ROOT

DATA_RAW = ROOT / "data" / "raw"

DATA_PROCESSED = ROOT / "data" / "processed"

FIGURES = ROOT / "figures"

# Set scope in years of available data
YEAR_MIN = 2015
YEAR_MAX = 2023
```

```python
# Verify all paths
print("ROOT:", ROOT)
print("RAW DATA:", DATA_RAW)
print("PROCESSED DATA:", DATA_PROCESSED)
print("FIGURES:", FIGURES)
```

---
**Load DataFrame**
---

```python
# Load raw data
RAW_DEMOGRAPHIC_ASPECTS_FILE = DATA_RAW / "Demographic-aspects-2023.xlsx"
```

---
**Test if file can be found**
---

```python
# Create "if statement" to test for errors
if not RAW_DEMOGRAPHIC_ASPECTS_FILE.exists():
    raise FileNotFoundError
```

---
**Inspect DataFrame**
---

```python
# Set variable for wide format
demographics_raw_wide = pd.read_excel(RAW_DEMOGRAPHIC_ASPECTS_FILE, header=1)
```

```python
demographics_raw_wide.describe()
```

---
**Extract the indicators columns of the data set**
---

```python
indicator_columns = demographics_raw_wide["Key Demographic aspects"]

# Inspect columns
indicator_columns
```

---
**Create variables for both 'males' and 'females'
Note: the data is in wide format**
---

```python
# Inspect and filter the data into our two groups
males = demographics_raw_wide[demographics_raw_wide["Key Demographic aspects"] == "Males"]
females = demographics_raw_wide[demographics_raw_wide["Key Demographic aspects"] == "Females"]
```

---
**Calulcate year-to-year changes in population by sex**
---

```python
# Calculate the yearly change (slicing 2015 to 2023)
# Using.diff(axis=1) to subtract each year from the previous one
yearly_change_males_df = males.loc[:, 2015:2023].diff(axis=1)
yearly_change_females_df = females.loc[:, 2015:2023].diff(axis=1)
```

```python
yearly_change_males_array = yearly_change_males_df.iloc[0].values
yearly_change_females_array = yearly_change_females_df.iloc[0].values
```

---
Note: Annual change starts in 2016. The 2015 value is undefined because no 2014 baseline is available.
2015 has been removed in the cell below using .droplevel(1)
---

```python
# Combine them into one comparison table
sex_comparison = pd.concat([yearly_change_males_df, yearly_change_females_df], keys=["Males", "Females"])

# Clean up: drop the extra index level
sex_comparison = sex_comparison.droplevel(1)

# Convert values and assign new variables
sex_comparison.columns = sex_comparison.columns.astype(int)
```

```python
# Variables to plor year values on chart
years = sex_comparison.columns
yearly_change_males_values = sex_comparison.loc["Males"].values
yearly_change_females_values = sex_comparison.loc["Females"].values

years_index = years
years_list = list(years)
last_year = years_list[-1]
```

---
# Creation of final visualization
**Plotting, showing image and saving figure to the figure directory**
---

---
**Create variables to plot abs. values on line graph**
---

```python
males_abs_yr_2018 = float(males[2018].iloc[0])
males_abs_yr_2020 = float(males[2020].iloc[0])
males_abs_yr_2023 = float(males[2023].iloc[0])
```

```python
females_abs_yr_2018 = float(females[2018].iloc[0])
females_abs_yr_2020 = float(females[2020].iloc[0])
females_abs_yr_2023 = float(females[2023].iloc[0])
```

```python
# Create plot dimension
plt.figure(figsize=(15, 10))

# Plot
plt.title("Male and Female Population Change in Aruba (2016–2023)")
plt.xlabel("Year")
plt.ylabel("Annual Growth / Decline (Absolute)")
plt.grid(True, alpha=0.3)

plt.plot(years, yearly_change_males_array, marker="o", label="Males")
plt.plot(years, yearly_change_females_array, marker="o", label="Females")

# Zero baseline
plt.axhline(0, color="black", linestyle="--", linewidth=1)

plt.text(2018, yearly_change_males_array[3] + 10, f'{males_abs_yr_2018:,.0f}')
plt.text(2020 + 0.1, yearly_change_males_array[5] + 10, f'{males_abs_yr_2020:,.0f}')
plt.text(2023, yearly_change_males_array[8] + 12, f'{males_abs_yr_2023:,.0f}')

plt.text(2018, yearly_change_females_array[3] + 10, f'{females_abs_yr_2018:,.0f}')
plt.text(2020 + 0.1, yearly_change_females_array[5] + 10, f'{females_abs_yr_2020:,.0f}')
plt.text(2023, yearly_change_females_array[8] + 10, f'{females_abs_yr_2023:,.0f}')

# Save figure
plt.savefig(FIGURES / "cbs_demographic_changes-2015-2023.png", dpi=300)

plt.show()
```
