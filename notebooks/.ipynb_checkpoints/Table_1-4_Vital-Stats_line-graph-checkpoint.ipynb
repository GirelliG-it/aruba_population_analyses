```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl

# Import Class
from pathlib import Path

# Load DataFrame
df = pd.read_excel("/home/ggirelli/Documents/DataAnalysis/Projects/Capstone_1/data/raw/Table-1.4-Vital-Statistics-1.xlsx", header=None)
```

```python
# Test DataFrame
df
```

```python
# Isolate indicators
indicator = df.iloc[3:9]
indicator
```

```python
# Iasolate year series
df_year = df.iloc[1]
df_yr_clean = df_year[1:].ffill()
df_yr_clean
```

```python
# Isolate sex series
df_sex = df.iloc[2]
df_sex_clean = df_sex[1:]
df_sex_clean
```

```python
# Use zip() to merge Series
df_yr_sex = list(zip(df_yr_clean, df_sex_clean))
df_yr_sex

# Create a title for first column to match index of indicator Series
final = ["Indicator"] + (df_yr_sex)
len(final)
```

```python
indicator.columns = final
indicator.head()
indicator = indicator.set_index("Indicator")
```

```python
#indicator = indicator.set_index("Indicator")
indicator.head()
```

```python
tidy = indicator.stack().reset_index()
tidy
```

```python
tidy[["Year", "Sex"]] = pd.DataFrame(tidy["level_1"].tolist(), index=tidy.index)

tidy
```

```python
tidy = tidy.rename(columns={0: "Value"})

tidy = tidy.drop(columns=["level_1"])

tidy
```

```python
tidy.shape
```

```python
tidy = tidy[["Indicator", "Year", "Sex", "Value"]]


tidy.to_csv("test.csv", index=False)
tidy.head()
```

```python
tidy["Indicator"] = tidy["Indicator"].str.strip()

tidy["Indicator"].unique()
```

```python
drivers = tidy[tidy["Indicator"].isin(["Natural increase", "Net migration"])]

drivers.head()
```

```python
drivers.shape
```

```python
drivers_pivot = drivers.pivot_table(
    index="Year",
    columns="Indicator",
    values="Value",
    aggfunc="sum"
)

drivers_pivot
```

```python
ax = drivers_pivot.plot(marker="o", figsize=(10, 6))

ax.set_title("Drivers of Population Change in Aruba (2015-2023")
ax.set_xlabel("Year")
ax.set_ylabel("Population change (people)")
ax.grid(True)

plt.show()
```
